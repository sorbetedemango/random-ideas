```python
import random
import matplotlib.pyplot as plt
import math


 
def prueba(x):
    a = random.randint(0,99)
    if a >= 75:
        return 0
    else:
        return 2 * x
        
    
bote = 100
#si partido no es 0, apuesta será interpretado como el porcentaje del bote actual que se apostará
def pronostico(bote,apuesta,partido):
    total = 0
    lista = []
    for j in range(100):
        bote2 = bote
        for i in range(100):
            if bote2 <= 0:
                lista.append(bote2)
                continue 
            if partido == 1:
                apuesta_actual = min(bote2,math.ceil(bote2 * apuesta))
            else:
                apuesta_actual = min(bote2,apuesta)
            bote2 -= apuesta_actual
            bote2 += prueba(apuesta_actual)
            lista.append(bote2)
        total += bote2
    media = total/100
    p = 0
    listamedias = []
    listariesgo = []
    for i in range(100):
        suma = 0
        a = p
        lista2 = []
        for j in range(100):
            suma += lista[a]
            lista2.append(lista[a])
            a += 100
        listariesgo.append(lista2.count(0) / 100)
        media2 = suma/100
        listamedias.append(media2)
        p+=1
        
        
    return media,listamedias,listariesgo

def pronostico2(bote,apuesta):
    total = 0
    lista = []
    for j in range(100):
        bote2 = bote
        for i in range(100):
            if bote2 <= 0:
                lista.append(0)
            else:
                bote2 -= apuesta
                bote2 += prueba(apuesta)
                lista.append(bote2)
        total += bote2
    media = total/100
    p = 0
    listamedias = []
    listariesgo = []
    for i in range(100):
        suma = 0
        a = p
        lista2 = []
        for j in range(100):
            suma += lista[a]
            lista2.append(lista[a])
            a += 100
        listariesgo.append(lista2.count(0) / 100)
        media2 = suma/100
        listamedias.append(media2)
        
        p+=1
        
        
        
    return media,listamedias,listariesgo
def comparar(x,y):
    a1 = 0
    b1 = 0
    for i in range(100):
        a,z,y = pronostico(100,x,1)
        b,z,y = pronostico(100,y,1)
        a = float(a)
        b = float(b)
        if a > b:
            a1+= 1
        else:
            b1+= 1
    return(f'{x} porcentaje gana un {a1}% de las veces y {y} gana un {b1}% de las veces')
    


    
print(pronostico(100,100,0))

```
